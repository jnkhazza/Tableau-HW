```python
import pandas as pd
```

```python
citi_data = pd.read_csv('july_2013_citibike_data.csv', encoding='utf-8')
```

```python
citi_df = pd.DataFrame(citi_data)
```

```python
citi_df['gender'] = citi_df['gender'].astype(str)
```

```python
citi_df.dtypes
```

```python
citi_df['gender'] = citi_df['gender'].replace(['0','1','2'],['Unknown','Male','Female'])
```

```python
citi_df.head()
```

```python
start_counts = citi_df['start station name'].value_counts().to_dict()
end_counts = citi_df['end station name'].value_counts().to_dict()
```

```python
start_df = pd.DataFrame(list(start_counts.items()),columns = ['start station name','frequency']) 
end_df = pd.DataFrame(list(end_counts.items()),columns = ['end station name','frequency'])
# start_dfx = pd.DataFrame(list(start_counts.items()),columns = ['station name','frequency']) 
# end_dfx = pd.DataFrame(list(end_counts.items()),columns = ['station name','frequency'])
start_df["start/end"] = "Start"
end_df["start/end"] = "End"
# start_dfx["start/end"] = "Start"
# end_dfx["start/end"] = "End"
start_df
```

```python
start_stop_df = start_df.append(end_df, ignore_index=True)
start_stop_df
```

```python
end_df.to_csv('Clean/end_df.csv', index = False)
start_df.to_csv('Clean/start_df.csv', index = False)
citi_df.to_csv('Clean/citi_df.csv', index = False)
start_stop_df.to_csv('Clean/start_stop_df.csv', index = False)
```
